### melt()

`pandas.melt(dataset)`</br>

`dataset.melt(dataset)`

Gives us a direct comparison of each variable-value pair in the dataset

```python
visits.melt()   # melt the dataset
```

```python
scores.melt()   # melt the dataset
```

```python
# differnt notation
pandas.melt(scores)
```

### id_vars

`dataset.melt(id_vars=[''])`

Allows us to chose the id column(s)

```python
scores.melt(id_vars='student')  # set the id_var column
```

```python
visits.melt(id_vars='date') # set the id_var column
```

### var_name

`dataset.melt(var_name='')`

Allows us to name the variables column

```python
scores.melt(id_vars='student', var_name = 'subject')    # change the name of the variable column
```

```python
visits.melt(id_vars='date', var_name='page')    # change the name of the variable column
```

### value_name

`dataset.melt(value_name='')`

Allows us to name the values column

```python
scores.melt(id_vars='student', var_name = 'subject', value_name='score')    # change the names of both columns
```

```python
visits.melt(id_vars='date', var_name='page', value_name='visits')   # change the names of both columns
```
