In pandas, loc and iloc are two powerful indexers used to select data from a DataFrame

#### Loc
Label-based indexing

i.e You use row and column labels (names) to filter data.


```python
# import library
import pandas as pd
```

```python
df = pd.read_csv('datasets/HousePrice Data.csv')
df
```

```python
df.loc[:, 'Price_USD']
```

```python
df.loc[:5, ['House_ID', 'Area_sqft', 'Location', 'Bedrooms']]
```

```python
df.loc[1:5, ['House_ID', 'Area_sqft', 'Location', 'Bedrooms']]
```

```python
df.loc[[1, 3, 5, 7, 9], ['House_ID', 'Area_sqft', 'Location', 'Bedrooms']]
```

#### iloc
Integer-based indexing

i.e You use row and column positions (like slicing in NumPy)

```python
df
```

```python
df.iloc[:, 4]
```

```python
df.iloc[:, -1]
```

```python
df.head()
```

```python
df.iloc[:, 0:4]
```

```python
df.iloc[1:6, [1, 3]]
```
