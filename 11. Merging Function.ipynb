### Merging Function
It helps you combine two dataframes by matching the values in common columns

<img src="cover/merging.png" alt="mer" width="500">




#### import library

```python
import pandas as pd
```

#### Create a two DataFrame

```python
df1 = pd.DataFrame({'ID' : [1, 2, 3, 4],'Class' : ['A', 'B', 'C', 'D']})
df1
```

```python
df2 = pd.DataFrame({ 'ID': [1, 2, 3, 4], 'Section': ['X1', 'X2', 'X3', 'X4']})
df2
```

### Merge the two DataFrame

```python
pd.merge(df1, df2)
```

```python
pd.merge(df2, df1)
```

### on
We make a common here ID

```python
pd.merge(df1, df2, on='ID')
```

```python
df1
```

```python
df2
```

### What If we have different Item in ID

```python
df1 = pd.DataFrame({'ID' : [1, 2, 3, 4], 'Class' : ['A', 'B', 'C', 'D']})
df1
```

```python
df2 = pd.DataFrame({ 'ID': [1, 2, 3, 5], 'Section': ['X1', 'X2', 'X3', 'X4']})
df2
```

```python
pd.merge(df1, df2, on='ID')
```

#### how
inner: take common only (by default it's inner)

outer: take all either is common or not

right: it's take only right ID col

left: it's take on left ID col


#### inner

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on = "ID", how = 'inner')
```

#### Outer

```python
pd.merge(df1, df2, on = "ID", how = 'outer')
```

#### Left

```python
pd.merge(df1, df2, on = "ID", how = 'left')
```

#### right

```python
pd.merge(df1, df2, on = "ID", how = 'right')
```

#### Indicator
It tell how file are merge according to 'how' operation

```python
pd.merge(df1, df2, on = "ID", how = 'right', indicator=True)
```

```python
pd.merge(df1, df2, on = "ID", how = 'outer', indicator=True)
```

#### If we have no item similar in ID column in both dataframes

```python
df1 = pd.DataFrame({'ID' : [1, 2, 3, 4], 'Class' : ['A', 'B', 'C', 'D']})
df1
```

```python
df2 = pd.DataFrame({ 'ID': [5, 6, 7, 8], 'Section': ['X1', 'X2', 'X3', 'X4']})
df2
```

```python
pd.merge(df1, df2, on='ID')
```

#### left index = True right index = True
here it's make a different ID by default it's name ID with _X, _y

```python
pd.merge(df1, df2, left_index = True, right_index = True)
```

#### Suffixes
change the name of column

```python
pd.merge(df1, df2, left_index = True, right_index = True, suffixes=('_1', '_2'))
```
