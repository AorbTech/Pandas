### GroupBy

This function allows us to split our data into groups based on the values of one or more columns

##### Import Pandas

```python
import pandas
```

##### Import Datasets

```python
students = pandas.read_csv('./datasets/students.csv')
language_learners = pandas.read_csv('./datasets/language_learners.csv')
```

### groupby

`groupby(column(s))`

Splits data into groups

To access:
`.group`

```python
students
```

```python
students.groupby('sex').groups
```

```python
language_learners
```

```python
language_learners.groupby('language').groups
```

```python
language_learners.groupby('years_practiced').groups
```

```python
language_learners.groupby('level').groups
```

```python
language_learners.groupby('language')['years_practiced'].mean().round(2)
```

```python
language_learners.groupby('level')['years_practiced'].mean().round(2)
```

```python
language_learners.groupby('name')['language'].count()
```

### get_group

`groupby().get_group(n)`

Creates a dataframe from the desired group

```python
languages = language_learners.groupby('language')
languages.get_group('English')
```

```python
levels = language_learners.groupby('level')
levels.get_group('C2')
```

```python
levels.get_group('A2')
```

```python
people = language_learners.groupby('name')
people.get_group('Alice')
```

```python
people.get_group('David')
```
