#### Replace Value
Changing existing values in a DataFrame or Series with new values.

```python
# import library
import pandas as pd
import numpy as np
```

```python
### import dataset
df = pd.read_csv('datasets/HousePrice Data.csv')
df
```

```python
df.replace(to_replace='City Center', value= 'city')
```

```python
df.replace(to_replace=np.nan, value=0)
```

```python
df.replace(950.0, 980.0)
```

```python
df
```

```python
df['Bedrooms'] = df['Bedrooms'].replace(np.nan, 3)
df
```

### Regex
Regular expression used to replace the specific char in string

```python
df
```

```python
df.replace('r', 'RR', regex=True)
```

```python
df.replace("1", "one", regex=True)
```

```python
df.dtypes
```

```python
df.replace(1, "one", regex=True)
```

```python
df
```

```python
df.replace("1", "one", regex=False)
```

#### Methods
forward fill (replace previous value with forward value) 

backward fill (replace forward value with previous value)

```python
df.replace("Rural", method='ffill')  # forward fill
```

```python
df
```

```python
df.replace("H4", method='bfill')  # backward fill
```

```python
df.replace(np.nan, method='ffill') 
```

### Inplace

```python
df
```

```python
df.replace(np.nan, method='ffill', inplace=True) 
```
